```python
#Value Counts
#值计数是了解每个特征/变量有多少个单位的好方法。我们可以对“drive-wheels”列应用“value_counts”方法。请注意，“value_counts”方法仅适用于pandas Series，不适用于pandas DataFrame。因此，我们只需添加一个括号 df['drive-wheels']，而不是两个括号 df[['drive-wheels']]。
df['drive-wheels'].value_counts()
```

```python
#转化series为frame
df['drive-wheels'].value_counts().to_frame()
```

让我们重复上述步骤，但将结果保存到数据框“drive_wheels_counts”中，并将列“drive-wheels”重命名为“value_counts”。

```python
drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.reset_index(inplace=True) #设置列的索引
drive_wheels_counts=drive_wheels_counts.rename(columns={'drive-wheels': 'value_counts'}) #改变表格的名字
drive_wheels_counts
```

```python
#Now let's rename the index to 'drive-wheels':
drive_wheels_counts.index.name = 'drive-wheels'
drive_wheels_counts
```

```python
#从数据框 df 里取出 drive-wheels 这一列（一个 Series），然后用 unique() 方法，返回这一列中所有不重复的取值
df['drive-wheels'].unique()
```

```python
#选三列，做一个子表
df_group_one = df[['drive-wheels','body-style','price']]
# grouping results 按一列分组，只对 price 做聚合
#as_index：分组键是不是当成索引 True（默认）：分组列变成行索引。False：分组列保留为普通列（你前面用的写法）。
df_grouped = df_group_one.groupby(['drive-wheels'], as_index=False).agg({'price': 'mean'})
df_grouped
```

```python
# grouping results
df_gptest = df[['drive-wheels','body-style','price']]
#先按两列一起分组：每种 (drive-wheels, body-style) 组合一组。对每一组里 所有数值类型的列 计算平均值。
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
#创建透视表
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
#把缺失数据替换为0
grouped_pivot = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot
```

```python
# 假设前面已经有 grouped_pivot 这个 DataFrame

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 6))

# 只用 ax.pcolor 画一次热图
im = ax.pcolor(grouped_pivot, cmap='RdBu')

# 准备行列标签：
# 行：index（drive-wheels）
row_labels = grouped_pivot.index
# 列：columns（body-style；如果是 MultiIndex 就用 get_level_values(1)）
col_labels = grouped_pivot.columns  # 或 grouped_pivot.columns.get_level_values(1)

# 把刻度放在每个色块中心
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

# 贴上文字标签
ax.set_xticklabels(col_labels, minor=False)
ax.set_yticklabels(row_labels, minor=False)

# x 轴标签旋转 90 度
plt.xticks(rotation=90)

# 加颜色条（一定要用同一个 fig / im）
fig.colorbar(im, ax=ax)

plt.tight_layout()  # 防止文字被遮挡
plt.show()          # 只在最后调用一次
```

相关性

```python
#皮尔逊相关系数是函数“corr”的默认方法。和以前一样，我们可以计算“int64”或“float64”变量的皮尔逊相关系数。
df.select_dtypes(include=['number']).corr()
```

```python
#计算P值
#p 值为：0.001：有强有力的证据表明相关性显著。
#p 值为：0.05：有中等程度的证据表明相关性显著。
#p 值为：0.1：有较弱的证据表明相关性显著。
#p 值为：0.1：没有证据表明相关性显著。
##我们可以使用“scipy”库中的“stats”模块来获取此信息。
from scipy import stats
```

```python
#计算 p_value stats.pearsonr函数返回相关系数r和P值
pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])
print("The Pearson Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value) 
```
